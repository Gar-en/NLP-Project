```python
import pandas as pd
df_1 = pd.read_csv("E:/Gar/Project 5/MandatoryRequirement_1685111911413.csv")
print(df_1.head())
```

```python
df_2 = pd.read_csv("E:/Gar/Project 5/MandatoryRequirement_1685111852923.csv")
print(df_2.head())
```

```python
import json

# Конвертим столбец formData в список
df_1['formData'] = df_1['formData'].apply(json.loads)
```

```python
formData = df_1['formData'].apply(lambda obj: obj['data']).tolist()
df_1 = pd.DataFrame.from_dict(formData)
df_1.head()
```

```python
missing_values = df_1.isnull().sum() / len(df_1)
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(inplace=True)
missing_values
```

```python
df_2['formData'] = df_2['formData'].apply(json.loads)
formData = df_2['formData'].apply(lambda obj: obj['data']).tolist()
df_2 = pd.DataFrame.from_dict(formData)
```

```python
missing_values = df_2.isnull().sum() / len(df_2)
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(inplace=True)
missing_values
```

```python
df = pd.concat([df_1, df_2], ignore_index=True)
```

```python
missing_values = df.isnull().sum() / len(df)
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(inplace=True)
missing_values
```
